### TO FIND THE OUTLIERS BY COMPUTING Z SCORES

```python
mer['Z-score']=(mer['AF']-mer['AF'].mean())/mer['AF'].std()
```

```python
Out=mer[(mer['Z-score']>3)|(mer['Z-score']<-3)].index
```

```python
mer1=mer.drop(Out)
```

```python
mer1['Z-score']=(mer1['SR']-mer1['SR'].mean())/mer1['SR'].std()
```

```python
Out1=mer1[(mer1['Z-score']>3)|(mer1['Z-score']<-3)].index
```

```python
mer2=mer1.drop(Out1)
```

```python
mer2['Z-score']=(mer2['FSO']-mer2['FSO'].mean())/mer2['FSO'].std()
```

```python
Out2=mer2[(mer2['Z-score']>3)|(mer2['Z-score']<-3)].index
```

```python
mer3=mer2.drop(Out2)
```

```python
mer3['Z-score']=(mer3['TSO']-mer3['TSO'].mean())/mer3['TSO'].std()
```

```python
Out3=mer3[(mer3['Z-score']>3)|(mer3['Z-score']<-3)].index
```

```python
mer4=mer3.drop(Out3)
```

### Observe the distribution of values of certain attributes for each variant and check if the distributions are different

```python
import matplotlib.pyplot as plt
```

```python
X=a['AC']
legend=['VARIANT_A','VARIANT_B']
Y=b['AC']
plt.figure(figsize=(10,8))
plt.hist([X,Y],color=['red','blue'])
plt.legend(legend)
plt.show()
```

### Does the PH values of the variants significantly different from one another?

```python
import scipy.stats as stats
stats.ttest_ind(a['PH'],b['PH'])
```

NULL HYPOTHESIS:PH VALUES OF VARIANTS ARE SAME
ALTERNATE HYPOTHESIS:PH VALUES OF VARIANTS ARE SIGNIFICANTLY DIFFERENT
AS P VALUE IS LESS THAN 0.05 WE REJECT THE NULL HYPOTHESIS
HENCE PH VALUES OF  VARIANTS ARE SIGNIFICANTLY DIFFERENT

### Does a variant of one product has higher values of Q compared to the other?

```python
q=mer4.groupby(['VARIANT'])['Q'].max()
var=['A','B']
plt.figure(figsize=(12,10))
plt.bar(var,q)
plt.show()
```

Q VALUES OF VARIANT B ARE GREATER THAN VARIANT A
